```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("allergen_class.csv", encoding="iso-8859-1")
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df = df.dropna(how='any')
```

```python
new = df['AllergenClass'].str.split(',', expand=True) 
     
```

```python
df['MainAllergen']=new[0]
df['Allergen_2']=new[1]
df['Allergen_3']=new[2]
df['Allergen_4']=new[3]
df['Allergen_5']=new[4]
```

```python
df['Allergen Level']=np.nan
```

```python
df = df.loc[df['MainAllergen']!="None"]
```

```python
df = df.reset_index()
```

```python
for index,row in df.iterrows():
    if row['Allergen_2'] == None:
        df.at[index, 'Allergen Level'] = 1
    elif row['Allergen_3'] == None:
        df.at[index, 'Allergen Level'] = 2
    elif row['Allergen_4'] == None:
        df.at[index, 'Allergen Level'] = 3
    elif row['Allergen_5'] == None:
        df.at[index, 'Allergen Level'] = 4
    else:
        df.at[index, 'Allergen Level'] = 5
```
